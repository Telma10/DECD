```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

import os

data_path = 'https://raw.githubusercontent.com/telmafaustino/DECD/main/'

consumos_faturados_por_municipio_df = pd.read_csv(data_path + 'consumos-faturados-por-municipio-ultimos-10-anos.csv', index_col=False, sep=';')
consumos_faturados_por_municipio_df
```

```python
consumos_faturados_por_municipio_df['Ano'].describe()
consumos_faturados_por_municipio_df.info()
```

```python
consumos_faturados_por_municipio_df = consumos_faturados_por_municipio_df.loc[consumos_faturados_por_municipio_df['Ano'] >= 2022]
```

```python
consumos_faturados_por_municipio_df['Ano'].describe()
consumos_faturados_por_municipio_df.info()
```

```python
consumos_faturados_por_municipio_df['Distrito'] = consumos_faturados_por_municipio_df['Distrito'].str.capitalize()
consumos_faturados_por_municipio_df['Freguesia'] = consumos_faturados_por_municipio_df['Freguesia'].str.capitalize()
consumos_faturados_por_municipio_df
```

```python
consumos_faturados_por_municipio_df.to_csv('consumos_faturados_por_municipio_df_processed.csv')
```
