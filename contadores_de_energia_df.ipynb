```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

import os

data_path = 'https://raw.githubusercontent.com/telmafaustino/DECD/main/'

contadores_de_energia_df = pd.read_csv(data_path + 'contadores-de-energia.csv', index_col=False, sep=';')
contadores_de_energia_df
```

```python
contadores_de_energia_df.info()
contadores_de_energia_df['Ano'].describe()
```

```python
contadores_de_energia_df['inclui_contador_inteligente_flag'] = contadores_de_energia_df['Inclui contador inteligente'].apply(lambda x: 1 if x == 'SIM' else 0)
contadores_de_energia_df['inclui_mobilidade_eletrica_flag'] = contadores_de_energia_df['Contrato Ativo'].apply(lambda x: 1 if x == 'SIM' else 0)
contadores_de_energia_df
```
